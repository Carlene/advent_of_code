```python
''' 1. check first character
    2. if f, take the lower half of 128 rows (0-63). if b, upper half (64-127)
    3. save which half has been taken
    4. repeat 7 times, for the first seven characters in the boarding pass
    5. number that is left is the row
    
    5. check 8th character
    6. if l, take the lower half of 7 rows (4-7). if r, upper half (0-3)
    7. repeat for the remaining characters of the boarding pass
    8. number that is left is the column
    
    9. multiply row number by 8
    10. add multiplied row with the column
    11. return that number
'''
```

```python
def find_row(boarding_pass):
    row_min = 0
    row_max = 127
    difference = row_max - row_min
    front = 'F'
    back = 'B'
    
    for character in boarding_pass[:7]:
        if character == front:
            row_max -= ((difference // 2) + (difference % 2))
            current_row = row_max
        elif character == back:
            row_min += ((difference // 2) + (difference % 2))
            current_row = row_min
        else:
            current_row = False
            
        difference //= 2
            
    return current_row
```

```python
def find_column(boarding_pass):
    column_min = 0
    column_max = 7
    difference = column_max - column_min
    lower_half = 'L'
    upper_half = 'R'
    
    for character in boarding_pass[7:]:
        if character == lower_half:
            column_max -= ((difference // 2) + (difference % 2))
            current_column = column_max
        elif character == upper_half:
            column_min += ((difference // 2) + (difference % 2))
            current_column = column_min
        else:
            current_column = False
            
        difference //= 2
            
    return current_column
        
```

```python
def test(actual, expected):
    if actual != expected:
        print(f"I expected {expected}, but got {actual}")
    else:
        print("Looks good to me!")
```

```python
test(find_row('BFFFBBFRRR'), 70)
test(find_row('FFFBBBFRRR'), 14)
test(find_row('BBFFBBFRLL'), 102)
```

```python
test(find_column('BFFFBBFRRR'), 7)
test(find_column('FFFBBBFRRR'), 7)
test(find_column('BBFFBBFRLL'), 4)
```

```python
#multiply row by 8, add column for seat id
```

```python
def find_seat_ID(boarding_pass):
    column = find_column(boarding_pass)
    row = find_row(boarding_pass)
    
    return (row * 8) + column
```

```python
test(find_seat_ID('BFFFBBFRRR'), 567)
test(find_seat_ID('FFFBBBFRRR'), 119)
test(find_seat_ID('BBFFBBFRLL'), 820)
```

```python
def find_highest_seat_ID(list_of_boarding_passes):
    highest_seat_ID = 0
    
    for boarding_pass in list_of_boarding_passes:
        seat_ID = find_seat_ID(boarding_pass)
        if seat_ID > highest_seat_ID:
            highest_seat_ID = seat_ID
    
    return highest_seat_ID
```
