```python
#this skips rows, has to go three across on every row

def slope_movement(toboggan_map):
    slope_horizontal = 3
    start = 0
    tree_count = 0

    for i, area in enumerate(toboggan_map[1:]):
        start = (start + slope_horizontal) % len(area)
        if area[start] == '#':
            tree_count += 1

    return tree_count
```
