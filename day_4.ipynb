```python
import re
```

```python
def remove_new_lines(text):
    text = re.split("\n\n", text)
    
    for i, passport in enumerate(text):
        passport = passport.replace("\n", " ")
        text[i] = passport
    
    return text
```

```python
def make_list_of_lists_of_passports(unorganized_passports):
    grouped_lists_of_passports = []

    for passport in unorganized_passports:
        passport = passport.split(" ")
        grouped_lists_of_passports.append(passport)
        
    return grouped_lists_of_passports
```

```python
def make_passport_field_dictionary(text):
    unorganized_passports = remove_new_lines(text)
    grouped_lists_of_passports = make_list_of_lists_of_passports(unorganized_passports)
    
    list_of_passport_field_dictionaries = []

    for passport in grouped_lists_of_passports:
        passport_field_dictionary = {}
        count = 0
        for field in passport:
            colon = field.find(':')
            key = field[0:colon]
            value = field[(colon+1):]
            
            passport_field_dictionary[key] = value
            count+=1
            
            if count == (len(passport)):
                list_of_passport_field_dictionaries.append(passport_field_dictionary)
                passport_field_dictionary = {}
                count = 0
                
    return list_of_passport_field_dictionaries
```

```python
def test(actual, expected):
    if actual == expected:
        print("Seems OK to me!")
    else:
        print(f"I expected {expected} but got {actual}")
```
